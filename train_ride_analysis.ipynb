# find out the count of Route 

```python
df['Route']= df.apply(lambda x: " → ".join(sorted([x['Departure Station'], x['Arrival Destination']])), axis=1)
route_count = df['Route'].value_counts()
route_count
```

# Top 10 Most Popular Routes

```python
top_routes = route_count.head(10)
plt.figure(figsize=(10,8))
#top_routes.plot(kind='bar', color="Skyblue")
ax = sns.barplot(data=top_routes, color="Skyblue")
plt.title("TOp 10 Most Popular Routes", fontsize=16)
plt.ylabel('Number of Trips', fontsize=10)
plt.xlabel("Routes", fontsize=10)
for x in ax.containers:
    ax.bar_label(x)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('top_routes.png')
plt.show()
```

```python
# Group by hour of departure and count trips
df['Departure Time'] = pd.to_datetime(df['Departure Time'], format='%H:%M:%S').dt.time
df['Departure Hour'] = df['Departure Time'].apply(lambda x: x.hour)

# Group by hour of departure and count trips
peak_times = df['Departure Hour'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
peak_times.plot(kind='bar', color='salmon')

# Add titles and labels
plt.title('Peak Travel Times by Hour of Departure', fontsize=16)
plt.xlabel('Hour of Day', fontsize=12)
plt.ylabel('Number of Trips', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('travel_times.png')
plt.show()
```

```python
revenue_by_type_class=df.groupby(['Ticket Class', 'Ticket Type'])['Price'].sum().unstack()
ax = revenue_by_type_class.plot(kind='bar', stacked=False, figsize=(10,6), colormap='viridis')
for i in ax.containers:
    ax.bar_label(i)
plt.title('Revenue by Ticket Type And Ticket Class')
plt.xlabel('Ticket Class', fontsize=12)
plt.ylabel('TOtal Revenue', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('ticket_class_type.png')
plt.show()
```

```python
df['Actual Arrival Time']=df['Actual Arrival Time'].fillna(df['Arrival Time'])
```

```python
df['Actual Arrival Time'] = pd.to_datetime(df['Actual Arrival Time'], format='%H:%M:%S', errors='coerce')
df['Arrival Time'] = pd.to_datetime(df['Arrival Time'], format='%H:%M:%S', errors='coerce')
df['Delay Minutes'] = (df['Actual Arrival Time'] - df['Arrival Time']).dt.total_seconds()/60
df['On_Time']=df['Delay Minutes']<=10
```

```python
df.tail()
```

```python
on_time_counts= df['On_Time'].value_counts()
plt.figure(figsize=(6, 5))
on_time_counts.plot(
    kind='pie',
    labels=['On-Time', 'Delayed'],  # Adjust labels based on your data
    autopct='%1.1f%%',  # Show percentages
    colors=['skyblue', 'salmon'],  # Colors for each slice
    startangle=90,  # Rotate the chart for better visibility
    explode=(0.1, 0)  # Slightly separate the 'On-Time' slice
)
plt.title('On-Time Performance', fontsize=16)
plt.ylabel('')  # Remove default y-axis label
plt.tight_layout()
plt.savefig('on-time_per.png')
plt.show()
```

```python
# Filter only delayed trips
delayed_trips = df[df['On_Time'] == False]

# Count reasons for delay
delay_reasons = delayed_trips['Reason for Delay'].value_counts()
print(delay_reasons)

# Bar chart
plt.figure(figsize=(10, 6))
delay_reasons.plot(kind='bar', color='salmon')

# Add titles and labels
plt.title('Main Contributing Factors for Delays', fontsize=16)
plt.xlabel('Reason for Delay', fontsize=12)
plt.ylabel('Number of Delays', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('factor_of_delay.png')
plt.show()
```
